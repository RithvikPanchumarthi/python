```python
#IF STATEMENT

if 25>10:
  print("Yes, It Worked")
```

```python
#IF with Else STATEMENT

if 25<10:
  print("Yes, It Worked")
else:
  print("No, It wasn't worked")
```

```python
if 2<10:
  print("Yes, It Worked")
else:
  print("No, It wasn't worked")
```

```python
if 25<10:
  print("Yes, It Worked")
elif 25<30:
  print("Yes, elif Worked")
else:
  print("No, It wasn't worked")
```

```python
if (25<10) or (23<25):
  print("Yes, It Worked")
elif 25<30:
  print("Yes, elif Worked")
else:
  print("No, It wasn't worked")
```

```python
#Line Comprehension

print("It worked") if 10>30 else ("It didn't worked")
```

```python
#Nested If

if (25<10) or (23<25):
  print("Yes, It Worked")
  if 10<25:
    print("Nested If Statement worked")
elif 25<30:
  print("Yes, elif Worked")
else:
  print("No, It wasn't worked")
```

#WHILE LOOPS

```python
number = 1

while number <5:
  print(number)
  number = number + 1
```

```python
number = 1

while number <5:
  print(number)
  if number == 3:
    break
  number = number + 1
```

```python
number = 0

while number <=5:
  print(number)
  if number == 6:
    break
  number = number + 1
else:
  print("No longer less than 5")
```

```python
number = 0

while number <=5:
  number = number + 1
  if number == 3:
    continue
  print(number)
else:
  print("No longer less than 5")
```
